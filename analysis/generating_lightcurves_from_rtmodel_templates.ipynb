# Static Binary Lens signals from RTModel Template

```python
from jasmine import RTModelTemplateForBinaryLightCurve as rtm2l
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
template_path = '[your_path]/TemplateLibrary.txt'
```

### Different lines, different intervals

```python
lines = [2, 20, 40]
intervals = [(300, 302), (300, 350), (300, 400)]
fig, axes = plt.subplots(1,3)

for line, interval, ax in zip(lines, intervals, axes.ravel()):
    rtmodel_template_two_lenses = rtm2l(template_line=line,
                                        path_to_template=template_path,
                                        input_peak_t1=interval[0],
                                        input_peak_t2=interval[1])
    mag, times = rtmodel_template_two_lenses.rtmodel_magnification_using_vbb()
    ax.scatter(times, mag, s=0.8, alpha=0.3)
fig.tight_layout()
plt.show()
```
